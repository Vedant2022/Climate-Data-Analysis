# Section 2: Functions and Modules (16 marks)  

Here, I will build a function that calculates the moving average for a given dataset and column. The moving averages are one of several statistical techniques for analyzing a sequence of data points by averaging the data in windows of various sizes so you can gain insight into time-series variations with less noise from brief fluctuations. I will also make a function to calculate the correlation between two numeric columns using the pandas .corr() method. Correlation is a measure of the strength and direction of the linear relationship shared between two variables.

```python
#Function for Moving Averages
def movingAverage(data, column, windowSize):
    average = []
    for i in range(len(data)):
        windows =   data[column].iloc[max(0,i - windowSize + 1):i + 1]

        movingAvg = sum(windows) / len(windows)
        
        average.append(movingAvg)

    return average
```

```python
#Function for finding correlation between columns
def correlationAnalysis(data, column1, column2):
    corr = data[[column1,column2]].corr()  #[1] Find reference at the end
    
    return corr
```

```python
# Using functions to calculate moving average and to know correlation between 'Temperature' and 'CO2 Emissions' columns

#Calling moving average function
movAvg = movingAverage(data,'Temperature', windowSize=3)
print(movAvg)
#Calculating the average of moving average
finalAvg = sum(movAvg)/len(movAvg)
print("Overall Average of Moving Average of Temperature is: ",finalAvg)
```

### Let's find correlation between 'Temperature' and 'CO2 Emissions'

```python
#Calling correlation analysis function
corrAnalysis = correlationAnalysis(data, 'Temperature', 'CO2 Emissions')
print(corrAnalysis)
```

I defined moving average and correlation analysis functions in this section. The moving average function was called with a window size of 3meaning that, for every data point, the average of that point and the two preceding ones is calculated. Then, I computed the moving averages average, which will give a general idea about the trend of the data over time.

For the correlation analysis, I called the function:  
'Temperature' and 'CO2 Emissions': The correlation of these two columns is -0.002775, which indicates a very weak negative correlation, nearly zero. This would imply that there is no significant linear relationship between Temperature and CO2 Emissions in the data set.

#### 1. Working with Line Chart

```python
def lineChart():
    plt.figure(figsize=(20,10))
    plt.plot(data['Temperature'],linestyle='-',color='lightgreen')
    plt.show() #[3] Find reference at the end

lineChart()
#The idea of code has been taken from external sources and no code is copied directly
```

The above line chart represents the entire dataset, which includes data points from different countries or regions. While such visualization gives a broad overview, it presents challenges for data analysts who seek temperature trends relevant to a specific country or region.

```python
countryName = 'India'
countryData = data[data['Country'] == countryName]

def lineChartSpecific():
    plt.figure(figsize=(12, 6))
    plt.plot(countryData['Temperature'], linestyle='-', color='red', label=f'Temperature in {countryName}')
    #Let's define the title x-label and y-label for better understanding
    plt.title(f'Temperature Variations in {countryName}')
    plt.xlabel('Observation Index')
    plt.ylabel('Temperature')
    plt.show()

lineChartSpecific()
```

#### 2. Working with Heatmap
The data needs to be prepared before the heatmap can be created; thus, we use a pivot table. Pivoting is a powerful data manipulation technique that transforms the dataset into a matrix-like structure; for the heatmap, we want to sum up temperature data where the aggregation function is average temperature by each country.

It has rearranged the data such that the rows represent different countries and the values in the matrix represent the mean temperature for that particular country.

```python
def heatmap():    
    dataPivot = data.pivot_table(values='Temperature', index='Country', aggfunc='mean') #[4] Find reference at the end
    plt.figure(figsize=(10, 8))
    sns.heatmap(dataPivot, annot=True, fmt=".2f", cmap='coolwarm', cbar_kws={'label': 'Average Temperature (C)'})
    plt.title('Heatmap of Average Temperature by Country')
    plt.xlabel('Country')
    plt.ylabel('Temperature')
    plt.show() #[5] Find references at the end

heatmap()
```

For specific countries we will need to filter the data

```python
specificCountries = ['India','United Kingdom','Sri Lanka','Bangladesh','China','Nepal','Uzbekistan']
dataFilter = data[data['Country'].isin(specificCountries)]

def heatmapSpecific():   
    dataPivotNew = dataFilter.pivot_table(values='Temperature', index='Country', aggfunc='mean')
    plt.figure(figsize=(8, 6))
    sns.heatmap(dataPivotNew, annot=True, cmap='coolwarm', cbar_kws={'label': 'Average Temperature (C)'})
    plt.title('Heatmap of Average Temperatures for Specific Countries')
    plt.ylabel('Country')
    plt.xlabel('Metrics')
    plt.show()

heatmapSpecific()
```

#### 3. Working with Histogram

```python
def histogram():
    plt.figure(figsize=(10, 6))
    sns.histplot(data['Temperature'], bins=20, color='blue')
    plt.title('Temperature Distribution')
    plt.xlabel('Temperature')
    plt.ylabel('Frequency')
    plt.show() #[6] Find references at the end

histogram()
```

#### 4. Working with Scatter Plot

```python
def scatterPlot():
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=dataFilter, x='Temperature', y='Humidity', hue='Country', palette='bright',marker='o')
    plt.title('Scatter Plot of Temperature vs Humidity')
    plt.xlabel('Temperature (C)')
    plt.ylabel('Humidity (%)')
    plt.show() #[7] Find reference at the end

scatterPlot()
```

In the above code, I created four critical visualizations to analyse the climate dataset, focusing on specific countries and their corresponding climate trends:

1. Heatmap: The pivot table was generated to calculate the average temperatures for the selected countries ('India','United Kingdom','Sri Lanka','Bangladesh','China','Nepal','Uzbekistan'). The heatmap provided a visual insight into those averages with a gradient color scheme where one could identify the temperature variations across the selected countries.

2. Line Chart: A line plot is drawn for temperature variation of a specific country over a period. This plot focused on temporal sequences including maximums and minimums of the temperature hence Serving as a clear understanding of temperature trends in that particular country.

3. Histogram: The histogram illustrates the distribution of temperature measurements over some countries, using a density curve to point out the concentration of the bulk of temperature observations. By focusing in on specific countries, the histogram allows for specific insight into their respective temperature distributions.

4. Scatter Plot: The relationship between temperature and humidity for a specific choice of countries is examined through a scatter plot. Each country was labelled with a different color so that one may easily distinguish among them. # Section 5: GUI Development (16 marks)  

This section aims to develop a user-friendly Graphical User Interface (GUI) application using Tkinter that should aim at making the process of data visualization and analysis for any dataset simpler.

1. Dataset Upload: Users can upload a CSV file, which will then can be viewed by users.

2. View Dataset: The first few rows of the uploaded dataset are displayed to the users for quick previewing.

3. Users can choose type of visualisations they want to see. Options include scatter plots, line charts, histograms and heatmaps, created using Seaborn and Matplotlib from the previous section.

#### Importing required libraries

```python
import tkinter as tk
from tkinter import ttk
from tkinter import filedialog
from tkinter import messagebox
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
```

BELOW CODE is not of my own and some parts of it has been taken from different sources like GitHub. One can find the references at the end.

```python
#[8] Find reference at end
#[9] Find reference at end
#[10] Find reference at end
class GUI:
    def __init__(self,root):
        self.root = root
        self.root.title("Coursework GUI")
        self.root.geometry("800x600")

        self.data = None
        self.canvas = None

        #Dataset Upload Button Design
        ttk.Button(root, text="Upload Dataset", command=self.uploadFile).pack(pady=80)
        
        #Dataset View Button Design
        ttk.Button(root, text="View Dataset", command=self.viewDataset).pack(pady=10)

        #Drop-down menu for graphs
        self.graphDisplay = tk.StringVar()
        self.graphDisplay.set("Select Chart Type")
        ttk.Label(root, text="Select Chart Type:").pack(pady=10)
        chart_dropdown = ttk.OptionMenu(root, self.graphDisplay,"Select" ,"Heatmap", "Line Chart", "Histogram", "Scatter Plot")
        chart_dropdown.pack(pady=10)

        # Button to generate chart
        ttk.Button(root, text="Generate Chart", command=self.generateChart).pack(pady=10)

    #Funtion for uploading a dataset
    def uploadFile(self):
        filePath = filedialog.askopenfilename(filetypes=[("CSV files", "*.csv")])
        if filePath:
            self.data = pd.read_csv(filePath)
            messagebox.showinfo("Success", "Dataset uploaded successfully!")
        else:
            messagebox.showwarning("Warning", "No file selected.")

    #Function for viewing a dataset
    def viewDataset(self):
        if self.data is None:
            messagebox.showerror("Error", "No dataset uploaded!")
            return

        # For the dataset preview
        dataPreview = tk.Toplevel(self.root)
        dataPreview.title("Dataset Preview")
        dataPreview.geometry("800x600")

        frame = tk.Frame(dataPreview)
        frame.pack(fill=tk.BOTH, expand=True)   

        # Making the window resizable
        dataPreview.resizable(True, True)

        # Displaying the dataset
        # text = tk.Text(data_preview, wrap="none")
        # text.insert(tk.END, self.data.to_string())

        text = tk.Text(frame, wrap="none")
        text.insert(tk.END, self.data.to_string())
        text.pack(side=tk.LEFT, fill=tk.BOTH, expand=True)

        #Verical Scrollbar
        verticalScrollbar = tk.Scrollbar(frame, orient=tk.VERTICAL, command=text.yview)
        verticalScrollbar.pack(side=tk.RIGHT, fill=tk.Y)
        text.configure(yscrollcommand=verticalScrollbar.set)

        #Horizontal Scrollbar
        horizontalScrollbar = tk.Scrollbar(dataPreview, orient=tk.HORIZONTAL, command=text.xview)
        horizontalScrollbar.pack(side=tk.BOTTOM, fill=tk.X)
        text.configure(xscrollcommand=horizontalScrollbar.set)

        # Window resizing
        dataPreview.grid_rowconfigure(0, weight=1)
        dataPreview.grid_columnconfigure(0, weight=1)

        text.pack(fill=tk.BOTH, expand=True)  # Expand the Text widget in both directions
        text.configure(state="disabled") #No editing in dataset

    #Function if no dataset is uploaded
    def filterData(self):
        """A placeholder function to handle filtering if needed."""
        if self.data is None:
            messagebox.showerror("Error", "No dataset uploaded!")
            return None
        return self.data
    

    #New funtion for heatmap
    def heatmapSpecificNew(self,fig, ax):
        try:
            sns.heatmap(self.data.corr(), annot=True, cmap="coolwarm", ax=ax)
            ax.set_title("Heatmap")
        except Exception as e:
            messagebox.showerror("Error", f"Failed to generate Heatmap: {e}")
    
    #New function for line chart
    def lineChartSpecificNew(self,fig, ax):
        try:
            numericColumns = self.data.select_dtypes(include=['float64', 'int64']).columns
            if len(numericColumns) < 2:
                messagebox.showerror("Error", "Dataset must have at least two numeric columns for a Line Chart!")
               
            self.data[numericColumns].plot(ax=ax)
            ax.set_title("Line Chart")
        except Exception as e:
            messagebox.showerror("Error", f"Failed to generate Line Chart: {e}")
            
    #New function for histogram
    def histogramNew(self,fig, ax):
        try:
            numericColumns = self.data.select_dtypes(include=['float64', 'int64']).columns
            if not numericColumns.any():
                messagebox.showerror("Error", "No numeric columns available for Histogram!")
                
            self.data[numericColumns[0]].plot(kind='hist', ax=ax, bins=20)
            ax.set_title("Histogram")
        except Exception as e:
            messagebox.showerror("Error", f"Failed to generate Histogram: {e}")
            
    #New function for scatter plot
    def scatterPlotNew(self,fig, ax):
        try:
            numericColumns = self.data.select_dtypes(include=['float64', 'int64']).columns
            if len(numericColumns) < 2:
                messagebox.showerror("Error", "Dataset must have at least two numeric columns for a Scatter Plot!")
                
            ax.scatter(self.data[numericColumns[0]], self.data[numericColumns[1]])
            ax.set_title("Scatter Plot")
            ax.set_xlabel(numericColumns[0])
            ax.set_ylabel(numericColumns[1])
        except Exception as e:
            messagebox.showerror("Error", f"Failed to generate Scatter Plot: {e}")
            

    #Function to view different graphs
    def generateChart(self):
        chart = self.graphDisplay.get()
        filtered_data = self.filterData()

        if filtered_data is None:
            return  # Exit if no data is loaded

        # Create a Matplotlib figure
        fig, ax = plt.subplots(figsize=(8, 6))
        plt.tight_layout()

        # Generate the appropriate chart
        if chart == "Heatmap":
            self.heatmapSpecificNew(fig, ax)
        elif chart == "Line Chart":
            self.lineChartSpecificNew(fig, ax)
        elif chart == "Histogram":
            self.histogramNew(fig, ax)
        elif chart == "Scatter Plot":
            self.scatterPlotNew(fig, ax)
        else:
            messagebox.showerror("Error", "Please select a valid chart type!")
            return


        #USING PREVIOUSLY BUILD FUNCTIONS OF VISUALISATIONS GENERATES ERROR AND HENCE WE HAVE TO EDIT IN THE FUNCTIONS 
        #INSTEAD I WILL BE CREATING COMPLETELY NEW FUNCTIONS FOR THE SAME
        #FIND THE FUNCTIONS ABOVE

        # Clear previous canvas and display the new chart
        if self.canvas:
            self.canvas.get_tk_widget().destroy()

        self.canvas = FigureCanvasTkAgg(fig, master=self.root)
        self.canvas.draw()
        self.canvas.get_tk_widget().pack(pady=10)


        

root = tk.Tk()
app = GUI(root)
root.mainloop()
```

I have successfully designed an GUI-based application usable for performing some fundamental data visualisation. The application allows the uploading of CSV datasets, shows a preview of the contents in a window that is both resizable and scrollable; it provides visualisations in heat maps, line charts, histogram, and scatter plot form. Every chart type is dealt with in its specific function. This GUI consists of buttons to upload and preview the dataset; a drop-down menu to select a chart type, and last but not least, canvas, which will show visualisation dynamically. Only the currently given chart is presented; previous charts get deleted to maintain a continuous visual flow by such deletion.
